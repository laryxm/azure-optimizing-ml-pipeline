# Optimizing an ML Pipeline in Azure

## Overview
This project is part of the Udacity Azure ML Nanodegree.
In this project, we build and optimize an Azure ML pipeline using the Python SDK and a provided Scikit-learn model.
This model is then compared to an Azure AutoML run.

## Summary

### The data set
The "Bank Marketing" data set contains information about direct marketing campaigns of a Portuguese banking institution.
In this project we seek to predict if the client would subscribe to a specif bank product ('yes') or not ('no'). The product studied was related to bank term deposit.
